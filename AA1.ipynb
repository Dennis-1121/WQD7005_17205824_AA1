Dataset Structure: 
CustomerID: Unique identifier for each customer. 
Age: Age of the customer. 
Gender: Gender of the customer. 
Location: Geographic location of the customer. 
MembershipLevel: Indicates the membership level (e.g., Bronze, Silver, Gold, 
Platinum). 
TotalPurchases: Total number of purchases made by the customer. 
TotalSpent: Total amount spent by the customer. 
FavoriteCategory: The category in which the customer most frequently shops (e.g., 
Electronics, Clothing, Home Goods). 
LastPurchaseDate: The date of the last purchase. 
[Additional Attributes]: Consider adding more attributes like customer's occupation, 
frequency of website visits, etc. 
Churn: Indicates whether the customer has stopped purchasing (1 for churned, 0 for 
active).

```python
# Get the first 1000 unique 'CustomerID' records
first_1000_unique_records = data['cust_id'].unique()[:1000]

# Select the rows corresponding to the first 1000 unique 'CustomerID' values
result_df = data[data['cust_id'].isin(first_1000_unique_records)]

print(result_df)
```

```python
# Write the DataFrame to a CSV file
#result_df.to_csv('output.csv', index=False)
```

```python
# Create a new column 'TotalPurchase' representing the number of records with the same 'CustomerID'
result_df['TotalPurchases'] = result_df.groupby('cust_id')['cust_id'].transform('count')

# Create a new column 'TotalSpend' representing the sum of 'Total' for each 'CustomerID'
result_df['TotalSpent'] = result_df.groupby('cust_id')['total'].transform('sum')

# Create a new column 'TotalDiscount' representing the sum of 'DiscountAmount' for each 'CustomerID'
result_df['TotalDiscount'] = result_df.groupby('cust_id')['discount_amount'].transform('sum')

# Create a new column 'FavoriteCategory' representing the favorite category for each 'CustomerID'
result_df['FavoriteCategory'] = result_df.groupby('cust_id')['category'].transform(lambda x: x.mode().iloc[0])

# Create a new column 'FavoriteCategory' representing the favorite category for each 'CustomerID'
result_df['FavoritePaymentMethod'] = result_df.groupby('cust_id')['payment_method'].transform(lambda x: x.mode().iloc[0])

# Convert 'PurchaseDate' to datetime format
result_df['order_date'] = pd.to_datetime(result_df['order_date'])

# Create a new column 'LastPurchaseDate' representing the date of the last record for each 'CustomerID'
result_df['LastPurchaseDate'] = result_df.groupby('cust_id')['order_date'].transform('max')
```

```python
# Create a new column 'MembershipLevel' based on 'TotalPurchases' quartiles
result_df['MembershipLevel'] = pd.qcut(result_df['TotalPurchases'], q=[0, 0.2, 0.4, 0.6, 1.0], labels=['Bronze', 'Silver', 'Gold', 'Platinum'])
```

```python
# Convert 'LastPurchasesDate' to datetime format
result_df['LastPurchaseDate'] = pd.to_datetime(result_df['LastPurchaseDate'])

# Find the maximum date in the dataset
max_date = result_df['LastPurchaseDate'].max()

# Calculate the difference in months
result_df['DateDifference'] = (max_date - result_df['LastPurchaseDate']).dt.days // 30

# Set 'Churn' based on the difference in months
result_df['Churn'] = np.where(result_df['DateDifference'] > 12, 1, 0)

# Drop the intermediate 'DateDifference' column if not needed
result_df = result_df.drop('DateDifference', axis=1)
```
